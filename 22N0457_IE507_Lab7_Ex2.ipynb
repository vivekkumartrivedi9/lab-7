```python
import numpy as np
import pandas as pd
!pip install -q pyomo
from pyomo.environ import *
!apt-get install -y -qq coinor-cbc
opt_cbc=SolverFactory('cbc')
```

```python
data=np.loadtxt('lab7_ex2_lp_ip_coef.txt',delimiter=',')
```

```python
model=ConcreteModel()
m=len(data[0,:])-1
n=len(data[:,0])-1
```

```python
row_indices=np.arange(m)
col_indices=np.arange(n)
```

```python
model.x=Var(row_indices,domain=Binary)
```

```python
model.objective=Objective(expr=sum(model.x[i]*data[0][i] for i in row_indices),sense=maximize)
```

```python
model.constraints=ConstraintList()
```

```python
for j in col_indices:
  model.constraints.add(expr=sum(data[j+1,i]*model.x[i] for i in row_indices) <=data[j+1,-1])
```

```python
model.pprint()
```

```python
result = opt_cbc.solve(model)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
  # display solution
print('\nObjective = ', model.objective())
result.solver.time
print(result)
```

```python
print('solver time=',result.solver.time)
```
